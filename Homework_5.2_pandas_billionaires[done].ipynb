## 4) How many male billionaires are there compared to the number of female billionares? What percent is that? Do they have a different average wealth?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.

```python
males = (df.gender == 'male')

len(df[males])
```

```python
females = (df.gender == 'female')

len(df[females])
```

```python
male_length = len(df[males])
female_length = len(df[females])

print((female_length/male_length)*100, "percent")
```

## 5) What is the most common source/type of wealth? Is it different between males and females?

## 9) How old is an average billionaire? How old are self-made billionaires  vs. non self-made billionaires? 

```python
#age
#group by type of wealth
#use if statement to check if self made

df.age.mean()
```

```python
selfmade = (df.selfmade == 'self-made')
inherited = (df.selfmade == 'inherited')

print("Self-made billionaires are" , df[selfmade].age.mean(), "and non self made ones are", df[inherited].age.mean(), "years old in average.")

```

## 10) Who are the youngest billionaires? Who are the oldest? Make a graph of the distribution of ages.

> **TIP:** You use `.plot()` to graph values in a column independently, but `.hist()` to draw a [histogram](https://www.mathsisfun.com/data/histograms.html) of the distribution of their values

```python
df[['name', 'age']].sort_values(by="age", ascending = True)
```

```python
df[['name', 'age']].sort_values(by="age", ascending = False)
```

```python
df[['name', 'age']].sort_values(by="age", ascending = True).plot(x = 'name', y = 'age').plot(kind='scatter')
```

## 11) Make a scatterplot of net worth compared to age

```python
df[['age','networthusbillion']].sort_values(by='networthusbillion', ascending = True).plot(x = 'networthusbillion', y = 'age').plot(kind='scatter')
```
